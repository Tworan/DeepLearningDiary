```python
import numpy as np
```

```python
from matplotlib import pyplot as plt

def softmax_temperature(a, temperature=0.5):
    a = np.abs(a)
    a = np.log(a) / temperature
    a = np.exp(a)
    a = a / np.sum(a)
    return a
```

## Temperature对softmax分布的影响
### Temperature越大则最后分布的熵越大，选择更具有随机性；Temperature越小则最后的熵越小，选择更稳定.
#### (另外，用Subplot作图)

```python
b = np.random.randn(10)
for i, temperature in enumerate([0.25, 0.5, 0.75, 1]):
    a = softmax_temperature(a=b, temperature=temperature)
    plt.subplot(2, 2, i+1)
    plt.hist(a)
    plt.title('Temperature:'+str(temperature))
```
