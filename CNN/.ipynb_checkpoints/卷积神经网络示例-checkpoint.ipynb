```python
import numpy as np
from tensorflow.keras import layers, models
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession
from tensorflow.compat.v1.keras.backend import set_session
from tensorflow.compat.v1 import ConfigProto, Session
import tensorflow.keras
config = ConfigProto(allow_soft_placement=True)

config.gpu_options.per_process_gpu_memory_fraction = 0.8
set_session(Session(config=config))
```

```python
def build_model():
    model = models.Sequential()
    model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(28,28,1,)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(10, activation='softmax'))
    model.summary()
    return model
```

```python
from tensorflow.keras.datasets import mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
def vectorizer(list):
    array = np.zeros((len(list), 10))
    for i, item in enumerate(list):
        array[i][item]=1
    return array
```

```python
Y_train = vectorizer(Y_train)
Y_test = vectorizer(Y_test)
```

```python
from sklearn.model_selection import train_test_split
train_data, val_data, train_labels, val_labels = train_test_split(X_train, Y_train)
```

```python
val_data = val_data.reshape(15000, 28, 28, 1)
train_data = train_data.reshape(len(train_data), 28, 28, 1)
X_test = X_test.reshape(len(X_test), 28, 28, 1)
```

```python
import datetime
from tensorflow.keras.callbacks import TensorBoard
log_dir = '/home/oneran/Mycodes/DeepLearning/log/Convert_Example_' + str(datetime.datetime.now())
callbacks_tensorflow = TensorBoard(log_dir=log_dir, histogram_freq=1)
model = build_model()
model.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(
    train_data,
    train_labels,
    validation_data=(val_data, val_labels),
    epochs=5,
    batch_size=64,
    callbacks=callbacks_tensorflow
)
```

```python
prediction = model.predict(X_test)
```
